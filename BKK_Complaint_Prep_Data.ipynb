```python
import pandas as pd
import requests
import math
```

**Find the total number of cases (round down)**

```python
r = requests.get("https://publicapi.traffy.in.th/share/teamchadchart/search?limit=1")
n_cases = r.json()["total"]
n_cases = math.floor(n_cases / 1000) * 1000
n_cases
```

```python
r = requests.get(f"https://publicapi.traffy.in.th/share/teamchadchart/search?limit=100&offset=0")
result = r.json()
df = pd.json_normalize(result["results"])
df
```

**Extract the data from API by using offest params**

```python
n = 0

df_output = pd.DataFrame()

while n <= n_cases:
  r = requests.get(f"https://publicapi.traffy.in.th/share/teamchadchart/search?limit=1000&offset={n}")
  result = r.json()
  df = pd.json_normalize(result["results"])
  try:
    print(f"https://publicapi.traffy.in.th/share/teamchadchart/search?limit=1000&offset={n}")
    n = n + 1000
    df_output = df_output.append(df, ignore_index = True)
  except:
    pass

# # add index as a new column
# df.reset_index(inplace=True)
# df = df.rename(columns = {'index':'id'})

df_output.columns = df_output.columns.str.replace('.', '_')
df_output.info()
```

**Add dataframe to Googlesheet**

```python
# authenticate
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

# create, and save df
from gspread_dataframe import set_with_dataframe
title = 'BKK_Complaints Summary'
#gc.create(title)  # if not exist
sheet = gc.open(title).sheet1
set_with_dataframe(sheet, df_output, include_index=False, include_column_header=True, resize=False)

print("Data updated successfully!!!")
```
